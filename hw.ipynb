```python
df['order_month'] = df['order_date'].dt.month
df['order_weekday'] = df['order_date'].dt.dayofweek 
```

```python
df['profit_margin'] = df['profit'] / df['sales']
```

```python
df['sales_per_qty'] = df['sales'] / df['quantity']
```

```python
df['discount_effect'] = df['sales'] * (1 - df['discount'])
```

```python
df['shipping_days'] = (df['ship_date'] - df['order_date']).dt.days
```

```python
df['high_discount'] = (df['discount'] > 0.2).astype(int)
```

```python
df['total_cost'] = df['sales'] - df['profit'] + df['shipping_cost']
```

```python
print(df[['order_month','order_weekday','profit_margin','sales_per_qty','discount_effect',
          'shipping_days','high_discount','total_cost']].head())
```

```python


Order date se order_month aur orderweekday features create kiye gaye,
jisse monthly aur weekly sales patterns ko analyse karna easy ho gaya.
 
Sales aur profit ka use karke profitmargin feature banaya gaya,
jo product aur order-wise profitability ko clearly show karta hai.

Sales per quantity feature create kiya gaya jisse har product ka
average unit price aur performance compare ki ja sake.

Discount effect feature add kiya gaya taaki discount ka
actual sales value par kya impact pad raha hai, yeh samjha ja sake.

Shipping days aur total cost features create karke
delivery efficiency aur overall order cost ka better analysis possible hua.
```
